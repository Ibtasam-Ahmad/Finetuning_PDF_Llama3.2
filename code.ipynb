````python
import os
import fitz  # PyMuPDF
from pathlib import Path

def convert_pdfs_to_images(pdf_dir):
    # Ensure the directory exists
    pdf_dir = Path(pdf_dir)
    output_root = Path("./AI_pdfs")
    output_root.mkdir(exist_ok=True)

    if not pdf_dir.is_dir():
        print(f"Error: The directory {pdf_dir} does not exist.")
        return

    # Iterate over all PDF files in the directory
    for pdf_file in pdf_dir.glob("*.pdf"):
        # Create a subfolder for this PDF inside ./AI_pdfs
        pdf_output_folder = output_root / pdf_file.stem
        pdf_output_folder.mkdir(exist_ok=True)

        # your-openai-api-key the PDF
        doc = fitz.your-openai-api-key(pdf_file)

        # Iterate over each page in the PDF
        for page_num in range(doc.page_count):
            page = doc.load_page(page_num)

            # Define the zoom factor for 300 DPI
            zoom = 300 / 72
            mat = fitz.Matrix(zoom, zoom)

            # Get the pixmap (image) of the page
            pix = page.get_pixmap(matrix=mat)

            # Create the image path in the subfolder
            image_path = pdf_output_folder / f"{pdf_file.stem}_page_{page_num + 1}.png"

            # Save the image
            pix.save(image_path)
            print(f"Saved: {image_path}")

        doc.close()

    print("PDF to image conversion completed.")

# Example usage
convert_pdfs_to_images(r'./AI_Train')
````

````python
import os
import openai
import json

# Set your openai API key
openai.api_key = api_key

# Remove the repeated prefix from filenames
def remove_repeated_prefix(filename, prefix):
    # If the filename starts with the prefix, remove the first occurrence of the prefix
    if filename.startswith(prefix):
        return filename[len(prefix):]
    return filename

# Get the normalized filenames by removing the repeated prefix
def get_normalized_filenames(folder_path, prefix_to_remove=None):
    files = os.listdir(folder_path)
    files = [f for f in files if os.path.isfile(os.path.join(folder_path, f))]
    
    # Normalize file names by removing the repeated prefix
    if prefix_to_remove:
        files = [remove_repeated_prefix(f, prefix_to_remove) for f in files]
    
    return set(files)

# Generate questions and answers based on file content using openai API
def generate_qa_from_text(text1, text2):
    prompt = f"Please generate at least 25 questions and answers combined based on the following text files:\n\n{text1, text2}"

    response = client.responses.create(
    model="gpt-4o-mini",
    input=[
        {
            "role": "developer",
            "content": prompt
        },
        {
            "role": "user",
            "content": {text1, text2},
        }
    ]
)

    # Extract the generated Q&A from the response
    qa_pairs = response.output_text
    return qa_pairs

# Compare two folders and process files if they are the same
def compare_and_process_folders(folder1, folder2, json_folder, prefix='raw_'):
    # Create the output folder if it doesn't exist
    os.makedirs(json_folder, exist_ok=True)

    # Get file names, normalize by removing the 'raw_' prefix from folder2
    files1 = get_normalized_filenames(folder1, prefix_to_remove=prefix)
    files2 = get_normalized_filenames(folder2, prefix_to_remove=prefix)

    # Compare
    if files1 == files2:
        print("✅ The folders contain the same files (ignoring repeated prefixes).")
        # Iterate over the files and generate QA pairs
        for filename in files1:
            # Construct the full file paths
            file1_path = os.path.join(folder1, filename)
            file2_path = os.path.join(folder2, filename)

            # Read content from both files (assuming they have the same name)
            with your-openai-api-key(file1_path, 'r') as file:
                text1 = file.read()

            with your-openai-api-key(file2_path, 'r') as file:
                text2 = file.read()

            # Generate Q&A from the file content
            qa_pairs = generate_qa_from_text(text1, text2)

            # Define the path for the output JSON file (same name as the original txt file)
            json_filename = os.path.splitext(filename)[0] + '.json'
            json_file_path = os.path.join(json_folder, json_filename)

            # Write the Q&A JSON output to the file
            with your-openai-api-key(json_file_path, 'w') as json_file:
                json.dump(qa_pairs, json_file, indent=2)

            print(f"✅ JSON file for '{filename}' created successfully.")
    else:
        print("❌ The folders have different files.")
        print("In folder1 but not in folder2:", files1 - files2)
        print("In folder2 but not in folder1:", files2 - files1)

# Example usage
folder1 = './img_data'
folder2 = './txt_data'
json_folder = './json_data'

compare_and_process_folders(folder1, folder2, json_folder)
````

````python
import os
import openai
import json
from pydantic import BaseModel
from typing import List, Optional  

# Define the output schema
class QAPair(BaseModel):
    question: str
    answer: str

class QAResponse(BaseModel):
    qas: List[QAPair]

# Set your openai API key
openai.api_key = api_key

def generate_qa_from_text(text: str) -> QAResponse:
    prompt = f"Please generate at least 25 questions and answers based on the following combined text:\n\n{text}"

    response = openai.ChatCompletion.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are an assistant."},
            {"role": "user", "content": prompt}
        ],
        response_format=QAResponse
    )

    return response.choices[0].message.parsed

def remove_repeated_prefix(filename: str, prefix: str) -> str:
    """Remove the repeated prefix from filenames."""
    return filename[len(prefix):] if filename.startswith(prefix) else filename

def get_normalized_filenames(folder_path: str, prefix_to_remove: Optional[str] = None) -> set:
    """Get normalized filenames by removing the repeated prefix."""
    files = os.listdir(folder_path)
    files = [f for f in files if os.path.isfile(os.path.join(folder_path, f))]
    if prefix_to_remove:
        files = [remove_repeated_prefix(f, prefix_to_remove) for f in files]
    return set(files)

def compare_and_process_folders(folder1: str, folder2: str, json_folder: str, prefix: str = 'raw_'):
    """Compare two folders and process matching files to generate structured Q&A JSON outputs."""
    os.makedirs(json_folder, exist_ok=True)

    files1 = get_normalized_filenames(folder1, prefix_to_remove=prefix)
    files2 = get_normalized_filenames(folder2, prefix_to_remove=prefix)

    if files1 == files2:
        print("✅ The folders contain the same files (ignoring repeated prefixes).")
        for filename in files1:
            file1_path = os.path.join(folder1, filename)
            file2_path = os.path.join(folder2, filename)

            with your-openai-api-key(file1_path, 'r') as file:
                text1 = file.read()

            with your-openai-api-key(file2_path, 'r') as file:
                text2 = file.read()

            # Combine both texts
            combined_text = text1 + "\n" + text2

            qa_response = generate_qa_from_text(combined_text)
            json_filename = os.path.splitext(filename)[0] + '.json'
            json_file_path = os.path.join(json_folder, json_filename)

            with your-openai-api-key(json_file_path, 'w') as json_file:
                json.dump(qa_response.dict(), json_file, indent=2)

            print(f"✅ JSON file for '{filename}' created successfully.")
    else:
        print("❌ The folders have different files.")
        print("In folder1 but not in folder2:", files1 - files2)
        print("In folder2 but not in folder1:", files2 - files1)

folder1 = './img_data'
folder2 = './txt_data'
json_folder = './json_data'

compare_and_process_folders(folder1, folder2, json_folder)
````

````python
import os
import pandas as pd

def merge_csv_files(input_folder, output_filename):
    # List to store dataframes
    dataframes = []
    
    # Loop through all files in the folder
    for file in os.listdir(input_folder):
        # Check if the file is a CSV file
        if file.endswith('.csv'):
            file_path = os.path.join(input_folder, file)
            try:
                # Read the CSV file into a DataFrame
                df = pd.read_csv(file_path)
                dataframes.append(df)
            except Exception as e:
                print(f"Error reading {file}: {e}")
    
    # Concatenate all dataframes vertically (axis=0)
    if dataframes:
        final_df = pd.concat(dataframes, ignore_index=True)
        
        # Save the merged dataframe to the final CSV file in the current directory
        final_df.to_csv(output_filename, index=False)
        print(f"Final dataset saved as {output_filename}")
    else:
        print("No CSV files found to merge.")

# Path to the folder containing CSV files
csv_folder = './csv_data'

# Name of the final merged dataset (will be saved in the current working directory)
final_output_filename = 'final_dataset.csv'  # No path, saved in the current working directory

# Merge the CSV files
merge_csv_files(csv_folder, final_output_filename)
````

````python
import os
import pandas as pd

def check_csv_columns(input_folder):
    # List to store names of CSV files with more than 2 columns
    problematic_files = []
    
    # Loop through all files in the folder
    for file in os.listdir(input_folder):
        if file.endswith('.csv'):
            file_path = os.path.join(input_folder, file)
            try:
                df = pd.read_csv(file_path)
                if len(df.columns) > 2:
                    problematic_files.append(file)
            except Exception as e:
                print(f"Error reading {file}: {e}")
    
    # Print the results
    if problematic_files:
        print("CSV files with more than 2 columns:")
        for filename in problematic_files:
            print(f"- {filename}")
    else:
        print("All CSV files have 2 or fewer columns.")

# Path to the folder containing CSV files
csv_folder = './csv_data'

# Run the check
check_csv_columns(csv_folder)
````
